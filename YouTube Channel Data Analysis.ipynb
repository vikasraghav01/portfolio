```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "YT_DS.csv"
data = pd.read_csv(file_path)
data.head()
```

```python
print(f"Shape of dataframe is: {data.shape}")
```

```python
data.describe()
```

```python
data.sort_values("Appx_Actual_Subscribers", ascending=False, inplace=True)
data[['YouTube Channel Name','Appx_Actual_Subscribers','YouTube_Channel_Link']].head()
```

```python
data.reset_index(inplace = True)
data.head()
```

Data Cleaning

```python
data.drop(['index'], axis=1, inplace=True)
data.columns
```

Statistical Info

```python
pd.DataFrame(data['Appx_Actual_Subscribers'].describe())
```

Visualization

```python
plt.figure(figsize = (7,6))
plt.title("Top 20 Best Data Science Youtube Channels\n", size = 15);
sns.barplot(y = "YouTube Channel Name", x = "Appx_Actual_Subscribers",data = data.head(20), palette="dark");
```

```python

plt.figure(figsize = (7,6))
plt.title("Top 20 Best Data Science Youtube Channels\n\n", size = 15);
sns.barplot(y = "YouTube Channel Name", x = "Appx_Actual_Subscribers",data = data.head(20), palette="bright")
plt.axvline(data['Appx_Actual_Subscribers'].mean(),linestyle = '--',color = 'k',linewidth = 2)
plt.text(13,-1, "Average Subscribers", fontsize = 13);
```

```python
above_1M = data[data['Above_1m'] == 1]
# (above_1M)
```

```python
plt.figure(figsize=(9,5))

plt.title("YT channels whose subscribers is above 1M\n", size = 15)
plots = sns.barplot(x = 'YouTube Channel Name',y = 'Appx_Actual_Subscribers',data = above_1M)

for bar in plots.patches:
    _ = plots.annotate(format(bar.get_height(), '.2f'), 
                       (bar.get_x() + bar.get_width() / 2, 
                        bar.get_height() - (bar.get_height()-3.1) / 2),
                        ha='center',
                        va='center',
                        size=15, xytext=(0, 0),
                        bbox=dict(boxstyle="round4, pad=0.6", fc="w", ec="k", lw=2),
                        textcoords='offset points')


plt.ylabel("Subscriber (in Million)")
plots.set_xticklabels(plots.get_xticklabels(),rotation = 30);
```

```python
def make_clickable(val):
    return '<a href="{}">{}</a>'.format(val,val)
above_1M[['YouTube Channel Name','YouTube_Channel_Link']].style.format({'YouTube_Channel_Link':make_clickable})
```

YT channels whose subscribers is b/w 200k and 1M

```python
bw_200k_and_1m = data[data['Between_200k_1m'] == 1]
plt.figure(figsize=(10,6))

plt.title("YT channels whose subscribers is b/w 200k and 1M\n", size = 15)
plots = sns.barplot(y = 'YouTube Channel Name',x = 'Appx_Actual_Subscribers',data = bw_200k_and_1m)

for bar in plots.patches:
        _ = plots.annotate(format(bar.get_width(), '.0f'), 
               (bar.get_width() / 2, 
                bar.get_y() + bar.get_height() / 2), ha='center', va='center',
               size=10, xytext=(0, 0),bbox=dict(fc="w", ec="k", lw=2),
               textcoords='offset points')

plt.ylabel("Channel Name")
plt.xlabel("\nSubscribers");
```

```python
bw_200k_and_1m[['YouTube Channel Name','YouTube_Channel_Link']].style.format({'YouTube_Channel_Link' : make_clickable})
```

From this dataset, most of the datascience Youtube channels has subscribers between 200k and 1M, which can also be shown below graph

```python
# taking log
data["log_Appx_Actual_Subscribers"] = np.log(data['Appx_Actual_Subscribers'])

# plotting
plt.rcParams["figure.figsize"] = [15, 20]


plt.title("Average subscribers of Data Science Channels on Youtube \n", size = 25);

sns.barplot(y = "YouTube Channel Name", x = "log_Appx_Actual_Subscribers",
            data = data, palette="cool_r")

plt.axvline(data['log_Appx_Actual_Subscribers'].mean(),linestyle = '--',color = 'k',linewidth = 2);
```
